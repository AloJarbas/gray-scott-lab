The goal is not a complete phase diagram on day one. The goal is a reproducible packet that shows how fixed seeding plus changing `(F, k)` produces visibly different reaction-diffusion fields.


## Model

The simulator advances the Gray-Scott system on a periodic square grid:

```text
u[t+1] = u[t] + D_u ∇²u - u v² + F (1 - u)
v[t+1] = v[t] + D_v ∇²v + u v² - (F + k) v
```

The central patch is seeded the same way for every run so the feed and kill changes are easier to read.


````python
from gray_scott_lab.analysis import CURATED_PRESETS, SCAN_FEEDS, SCAN_KILLS, scan_parameter_grid, study_presets

studies = study_presets(CURATED_PRESETS)
[(study.preset.name, study.metrics.active_fraction, study.metrics.edge_density) for study in studies]
````

The first useful split is between **how much chemistry survives** and **how sharp the interfaces stay**.

That is why the repo tracks both active fraction and edge density. One metric alone would blur sparse spots and dense labyrinths together.


````python
rows = scan_parameter_grid(SCAN_FEEDS, SCAN_KILLS)
max(rows, key=lambda row: row.metrics.edge_density)
````

## Reading the coarse scan

A coarse scan is enough for three day-one lessons:

1. Some `(F, k)` pairs barely sustain islands at all.
2. A middle band keeps a lot of activity while still holding sharp fronts.
3. A metric map is not the same thing as a complete phase diagram, but it is already better than a screenshot gallery with no quantitative spine.
