# Gray-Scott time evolution

This notebook is the slower companion to `reports/time-evolution-sidecar.md`.

The narrow question is the useful one: **how does one Gray-Scott regime grow into itself instead of only appearing as a final frame?**


## 1. Hold the chemistry fixed and watch the metrics move

This pass follows the curated `worm bands` preset with `F = 0.022` and `k = 0.051`.

The point is not a broad phase diagram. The point is to separate **filling in** from **front sharpening** inside one reproducible run.


```python
from gray_scott_lab.analysis import TIME_EVOLUTION_PRESET, study_time_evolution

study = study_time_evolution(TIME_EVOLUTION_PRESET)
[(point.step, round(point.metrics.active_fraction, 4), round(point.metrics.edge_density, 4)) for point in study.timeline[:8]]
```

## 2. Look for the split between occupancy and interface sharpness

If both metrics were just different spellings of the same thing, their most important moments would land at the same time. They do not.


```python
peak_active = max(study.timeline, key=lambda point: point.metrics.active_fraction)
peak_edge = max(study.timeline, key=lambda point: point.metrics.edge_density)
{
    'peak_active_step': peak_active.step,
    'peak_active': round(peak_active.metrics.active_fraction, 4),
    'peak_edge_step': peak_edge.step,
    'peak_edge': round(peak_edge.metrics.edge_density, 4),
}
```

That difference is the notebook version of the figure caption:

- active fraction tracks how much of the field stays chemically busy
- edge density tracks how sharp the interfaces stay while the fronts reorganize
- one preset can therefore carry both a growth story and a boundary story at the same time
